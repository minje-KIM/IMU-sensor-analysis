```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
  # train_seqs: 
      # 'dataset-outdoors1_512_16',
      # 'dataset-outdoors3_512_16',
      # 'dataset-outdoors5_512_16',
      # 'dataset-outdoors7_512_16',
      # 'dataset-outdoors8_512_16',
```

```python
file_paths = [
    '/data/minje/tumvi/Raw/dataset-outdoors1_512_16/mav0/imu0/data.csv',
    '/data/minje/tumvi/Raw/dataset-outdoors3_512_16/mav0/imu0/data.csv',
    '/data/minje/tumvi/Raw/dataset-outdoors5_512_16/mav0/imu0/data.csv',
    '/data/minje/tumvi/Raw/dataset-outdoors7_512_16/mav0/imu0/data.csv',
    '/data/minje/tumvi/Raw/dataset-outdoors8_512_16/mav0/imu0/data.csv',
]

# 데이터를 저장할 리스트 정의
data_list = []

# 파일을 읽어서 데이터 리스트에 추가
for file_path in file_paths:
    with open(file_path, 'r') as file:
        data = list(csv.reader(file, delimiter=','))
        data = np.array(data[1:], dtype=np.float32)
        data_list.append(data)
```

```python
print(" outdoors1 : ", data_list[0].shape, "\n",
     "outdoors1 : ", data_list[1].shape, "\n",
     "outdoors1 : ", data_list[2].shape, "\n",
     "outdoors1 : ", data_list[3].shape, "\n",
     "outdoors1 : ", data_list[4].shape, "\n")
```

```python
gyro_list = []
acc_list = []

for i in range(len(data_list)):

    # angular velocity
    
    wx = data_list[i][:,1]
    wy = data_list[i][:,2]
    wz = data_list[i][:,3]
    gyro = np.array([wx, wy, wz])
    gyro_list.append(gyro)
    
    # linear accleration
    
    ax = data_list[i][:,4]
    ay = data_list[i][:,5]
    az = data_list[i][:,6]
    acc = np.array([ax, ay, az])
    acc_list.append(acc)
    
```

```python
for i in range(5):
    
    plt.figure(figsize = (25,10))
    
    plt.subplot(3,2,1)
    plt.plot(acc_list[i][0], color = 'r')
    plt.title('acc_x', fontsize = 20)
    plt.ylim(-10,10)

    plt.subplot(3,2,2)
    plt.plot(gyro_list[i][0], color = 'r')
    plt.title('ang_x', fontsize = 20)
    plt.ylim(-3,3)

    plt.subplot(3,2,3)
    plt.plot(acc_list[i][1], color = 'g')
    plt.title('acc_y', fontsize = 20)
    plt.ylim(-5,5)

    plt.subplot(3,2,4)
    plt.plot(gyro_list[i][1], color = 'g')
    plt.title('ang_y', fontsize = 20)
    plt.ylim(-3,3)

    plt.subplot(3,2,5)
    plt.plot(acc_list[i][2], color = 'b')
    plt.title('acc_z', fontsize = 20)

    plt.subplot(3,2,6)
    plt.plot(gyro_list[i][2], color = 'b')
    plt.title('ang_z', fontsize = 20)
    plt.ylim(-3,3)

    plt.show()
```
